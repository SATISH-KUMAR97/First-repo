```python
df= pd.read_csv("/content/OFFENCE UNDER IT ACT.csv")
df.head()
```

```python
cs=df.iloc[0:26:2]
cs.head()
```

```python
cr=cs.iloc[0:24:1]
cr.head()
```

```python
cr.sort_values(['COMPUTER RELATED OFFENCE'], ascending=False, axis=0, inplace=True)
cr.head()
```

```python
plt.figure(figsize=(10,7))
plt.bar(np.arange(cr.shape[0]) -0.2,cr['COMPUTER RELATED OFFENCE'],width=0.2,color='m')
plt.bar(np.arange(cr.shape[0]) ,cr['PUBLICATION/TRANSMISSION OF OBSCENE/SEXUALLY EXPLICIT ACT IN ELECTRONIC FORM'],width=0.2,color='red')
plt.bar(np.arange(cr.shape[0]) + 0.2,cr['OTHER SECTIONS OF IT ACT'],width=0.2,color='blue')
plt.xticks(np.arange(cr.shape[0]), cr['STATE/UT'], rotation='vertical')
plt.show()
```
